```python
import tqdm
import pickle
import pandas as pd
from neo4j import GraphDatabase
```

```python
df_pw = pd.read_csv("pw.csv", delimiter=",")
```

```python
db = GraphDatabase.driver("neo4j://127.0.0.1:7687", auth=(df_pw["name"][0], df_pw["password"][0]))
```

```python
sess = db.session()
result = sess.run("MATCH (n:Transaction) Where date ({year: 2019, month: 7}) >n.date >= date({year: 2017, month: 8}) RETURN n LIMIT 25")
results = []
for r in result:
    results.append(r[0])
df_neo4j = pd.DataFrame(results)
```

```python
sess = db.session()
df_neo4j=pd.DataFrame([[1]], columns=[0])
count=0
while df_neo4j[0][0]>0:
    result = sess.run("MATCH (t:Transaction) WHERE t.date < date({year: 2017, month: 8}) OR t.date >= date({year: 2019, month: 7}) with t LIMIT 10000 DETACH DELETE t  RETURN count(*)")
    results = []
    for r in result:
        results.append(r[0])
    df_neo4j = pd.DataFrame(results)
    count+=1
    print(count)
```

```python
sess = db.session()
df_neo4j=pd.DataFrame([[1]], columns=[0])
count=0
while df_neo4j[0][0]>0:
    result = sess.run("MATCH (a:Address) where NOT (a)-[:RECEIVES]-(:Transaction) and not (a)-[:SENDS]-(:Transaction) with a limit 10000 Delete a")
    results = []
    for r in result:
        results.append(r[0])
    df_neo4j = pd.DataFrame(results)
    count+=1
    print(count)
```
